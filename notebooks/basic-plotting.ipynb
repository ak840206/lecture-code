Matplotlib
========

Matplotlib is a very popular plotting/visualization library in python. It is very popular for producing publication-quality charts and figures, even if the rest of the project is done in a language other than Python. In this notebook we will look at Matplotlib's model and some of its basic plotting mechanisms.

It is very common to import matplotlib's `pyplot` module as `plt`. We'll also need some data to visualize, so we're importing `numpy` as well.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Figures, Axes, Oh my
-------------------------------

Matplotlib use a notion of a 'Figure' as its main abstraction. This abstraction is presented as an object (in the object-oriented sense), so manipulating your figures and plots means interacting with your Figure object (and others, as we'll see). You can create a figure with the `plt.subplots()` function. Documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots

For our purposes today, we just need to know that with `plt.subplots()` we can create _a single figure_ with one or more _plots_. If we pass it no arguments, we get a `Figure` with a single subplot, otherwise we can provide it with the dimensions of how we would like our subplots to divide the figure.

In order to better demonstrate, let's create some data for us to plot:

```python
# Create some x values that would give us two periods of a standard sin wave
xs = np.arange(0,4*np.pi,0.1)

# y values for the standard sin
sin1 = np.sin(xs)

# y values for an amplified sin
sin2 = 2 * np.sin(xs)

# y values for an attenuated sin
sin3 = 0.5 * np.sin(xs)

# y values for a squashed sin
sin4 = np.sin(2*xs)
```

```python
# create a single figure with a single subplot:
#
# fig <- The figure that contains all of our subplots
#
# ax  <- the subplot (if there's only one)
#        or the list of subplots (if there's more than one)
fig, ax = plt.subplots()
```

And my Ax[e]
--------------------

Is we want to _plot_ something, we'll be plotting it on one of the 'axes'. Similarly, if we want to add labels we can do it on the `ax` in question:

```python
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
```

```python
# If we want to see the plot again in our notebook, we can ask for the _figure_:
fig
```

```python
ax.plot(xs,sin1)
fig
```

```python
ax.plot(xs,sin2)
fig
```

```python
ax.plot(xs,sin3)
ax.plot(xs,sin4)
fig
```

```python
data_set = [sin1, sin2, sin3, sin4]
colors = ['red','blue','yellow','magenta']
styles = ['dotted', 'dashdot', 'dashed', 'solid']

for i, d in enumerate(data_set):
    ax.plot(xs,d, linestyle=styles[i], color=colors[i])
    
fig
```

Multiple Subplots
-----

We can also plot each of these on a separate subplot:

```python
num_rows = 2
num_cols = 2

data_set = [sin1, sin2, sin3, sin4]

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols)

i = 0
for x in range(0,num_rows):
    for y in range (0, num_cols):
        axes[x,y].plot(xs,data_set[i])
        i += 1


for ax in axes.reshape(-1):
    ax.set_ylim([-2.5,2.5])
```

Legend
--------

We can also provide labels for each line:

```python
fig, ax = plt.subplots()

ax.plot(xs,sin3,linestyle='dashed',color='green',label='Wiggly boi')
ax.plot(xs,sin4,linestyle='dotted',color='green',label='Speedy boi')
ax.legend(loc='lower left')
```
